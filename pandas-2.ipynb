```python
# toss winner is match winner in percentage
new_dataset = ipl[['TossWinner','WinningTeam']]
sum_of_team = new_dataset.shape[0]
toss_and_mach_winner_team = 0
winner_team = 0
for i in range(new_dataset.shape[0]):
    if new_dataset['TossWinner'].iloc[i] == new_dataset['WinningTeam'].iloc[i]:
        toss_and_mach_winner_team += 1
    else:
        winner_team += 1

x = ['Toss and Mach Winner Team','WinningTeam']
y = [toss_and_mach_winner_team,winner_team]
fig,ax = plt.subplots()
ax.pie(y,labels=x)
plt.show()
```
